Explain the following concepts:

Supervised Learning: Supervised learning is when the training data fed to the algorithm includes the desired solutions, called labels.


Unsupervised Learning: Unsupervised learning is when the training data is unlabeled. As in, it does not include the desired solutions.


Online Learning: Online learning is when you train the system by incrementally feeding it data instances sequentially, either individually or in small groups called batches.


Batch Learning: Batched learning is when the system learns incrementally, using all the available data.


Model-based learning: Model-based learning is when you build a model of examples for the training, and then use those examples to make predictions.


Instance-based learning: This is when the system essentially memorizes the examples, and then learns to detect other cases that are similar ones it has memorized.


```python
#Step 1: Plot the noisy data and the polynomial
import matplotlib.pyplot as plt
import numpy as np

# Plot the noisy data:
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')

#Plot the polynomial
coefficients = np.polyfit(x.flatten(), y.flatten(), 2)
polynomial = np.poly1d(coefficients)
plt.plot(x, polynomial(x), 'bo')

```

```python
#Step 2: Plot the MSE vs m
import matplotlib.pyplot as plt
import numpy as np

noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
m = [1, 2, 3, 4, 5, 6, 7, 8]
mse = []
for i in m:
    coefficients = np.polyfit(x.flatten(), y.flatten(), i)
    polynomial = np.poly1d(coefficients)
    result = (sum((y - polynomial(x))**2))/number_of_samples
    mse.append(result[0])

print(mse)
plt.plot(m, mse, 'bo')

```

Explanation for Part 2: The degree that I would choose for m is 3. This is because the degrees of 3 through 8 have a relatively similar MSE, but the degree of 3 results in the most reasonable coefficients when using polyfit.
